```python
import pandas as pd
```

```python
df_test = pd.read_csv("testing_data.csv", index_col = 0)
```

```python
df_test.shape
```

```python
df_test.reset_index(drop = True, inplace = True)
```

```python
df_test = df_test.loc[df_test["game_next_goal"].isnull() == False, ]
```

```python
df_test.info()
```

```python
for col in df_test.columns:
    if df_test[col].dtype == "object":
        print(f"{col}: {df_test[col].unique()}")
        print()
```

```python
df_test["blue_handbrake"] = df_test["blue_handbrake"].astype("bool")
df_test["orange_handbrake"] = df_test["orange_handbrake"].astype("bool")

df_test["blue_ball_cam"] = df_test["blue_ball_cam"].astype("bool")
df_test["orange_ball_cam"] = df_test["orange_ball_cam"].astype("bool")

df_test["blue_boost_active"] = df_test["blue_boost_active"].astype("bool")
df_test["orange_boost_active"] = df_test["orange_boost_active"].astype("bool")
```

```python
df_test.loc[df_test["game_is_overtime"] == "True", "game_is_overtime"] = True
df_test.loc[df_test["game_is_overtime"].isnull(), "game_is_overtime"] = False
df_test.loc[df_test["game_is_overtime"] == "no_OT", "game_is_overtime"] = False
df_test["game_is_overtime"] = df_test["game_is_overtime"].astype("bool")

df_test["game_ball_has_been_hit"] = df_test["game_ball_has_been_hit"].astype("bool")
```

```python
df_test.loc[df_test["blue_dodge_active"]== "False", "blue_dodge_active"] = 0
df_test.loc[df_test["orange_dodge_active"] == "False", "orange_dodge_active"] = 0

df_test["blue_dodge_active"] = df_test["blue_dodge_active"].astype("float64")
df_test["orange_dodge_active"] = df_test["orange_dodge_active"].astype("float64")
```

```python
df_test.loc[df_test["blue_jump_active"]== "False", "blue_jump_active"] = 0
df_test.loc[df_test["orange_jump_active"] == "False", "orange_jump_active"] = 0

df_test["blue_jump_active"] = df_test["blue_jump_active"].astype("float64")
df_test["orange_jump_active"] = df_test["orange_jump_active"].astype("float64")
```

```python
df_test.loc[df_test["blue_double_jump_active"]== "False", "blue_double_jump_active"] = 0
df_test.loc[df_test["orange_double_jump_active"] == "False", "orange_double_jump_active"] = 0

df_test["blue_double_jump_active"] = df_test["blue_double_jump_active"].astype("float64")
df_test["orange_double_jump_active"] = df_test["orange_double_jump_active"].astype("float64")
```

```python
df_test.info()
```

```python
for col in df_test.columns:
    if df_test[col].dtype == "bool":
        print(f"{col}: {df_test[col].unique()}")
        print()
```
