# Model Comparison

Load in data

```python
import pandas as pd
%config InlineBackend.figure_formats = ['svg']
```

```python
X_accuracy = df_accuracy[["blue_ping", 
                          "blue_pos_x", 
                          "blue_pos_y", 
                          "blue_pos_z", 
                          "blue_vel_x", 
                          "blue_vel_y", 
                          "blue_vel_z", 
                          "blue_ang_vel_x", 
                          "blue_ang_vel_y", 
                          "blue_ang_vel_z", 
                          "blue_rot_x", 
                          "blue_rot_y", 
                          "blue_rot_z", 
                          "blue_jump_active", 
                          "blue_dodge_active", 
                          "blue_boost", 
                          "blue_possesion", 
                          "blue_dist_to_ball", 
                          "blue_behind_ball", 
                          "blue_dead", 
                          "orange_ping", 
                          "orange_pos_x", 
                          "orange_pos_y", 
                          "orange_pos_z", 
                          "orange_vel_x", 
                          "orange_vel_y", 
                          "orange_vel_z", 
                          "orange_ang_vel_x", 
                          "orange_ang_vel_y", 
                          "orange_ang_vel_z", 
                          "orange_rot_x", 
                          "orange_rot_y", 
                          "orange_rot_z", 
                          "orange_jump_active", 
                          "orange_dodge_active", 
                          "orange_boost", 
                          "orange_possesion", 
                          "orange_dist_to_ball", 
                          "orange_behind_ball", 
                          "orange_dead", 
                          "ball_pos_x", 
                          "ball_pos_y", 
                          "ball_pos_z", 
                          "ball_vel_x", 
                          "ball_vel_y", 
                          "ball_vel_z", 
                          "ball_ang_vel_x", 
                          "ball_ang_vel_y", 
                          "ball_ang_vel_z", 
                          "ball_dist_to_blue_goal", 
                          "ball_dist_to_orange_goal"]].copy()
X_accuracy.head()
```

Load in models

```python
import pickle
```

```python
clf_xgb = pickle.load(open("xgb_model2.model", "rb"))
clf_xgb
```

```python
clf_rf = pickle.load(open("rf_model.model", "rb"))
clf_rf
```

```python
clf_log = pickle.load(open("logistic_model.model", "rb"))
clf_log
```

### Compare Model Metrics

```python
df_accuracy["xgb_prediction_prob"] = clf_xgb.predict_proba(X_accuracy)[:, 1]
df_accuracy["xgb_prediction"] = clf_xgb.predict(X_accuracy)

df_accuracy["rf_prediction_prob"] = clf_rf.predict_proba(X_accuracy)[:, 1]
df_accuracy["rf_prediction"] = clf_rf.predict(X_accuracy)

df_accuracy["log_prediction_prob"] = clf_log.predict(X_accuracy)
df_accuracy["log_prediction"] = None
df_accuracy.loc[df_accuracy["log_prediction_prob"] < 0.5, "log_prediction"] = 0
df_accuracy.loc[df_accuracy["log_prediction_prob"] >= 0.5, "log_prediction"] = 1

df_accuracy.head()
```

Compare prediction accuracies

```python
cm_xgb = pd.crosstab(df_accuracy['game_next_goal'], df_accuracy['xgb_prediction'],
                     rownames=['Actual'], colnames=['Predicted'])
cm_xgb
```

```python
cm_rf = pd.crosstab(df_accuracy['game_next_goal'], df_accuracy['rf_prediction'],
                     rownames=['Actual'], colnames=['Predicted'])
cm_rf
```

```python
cm_log = pd.crosstab(df_accuracy['game_next_goal'], df_accuracy['log_prediction'],
                     rownames=['Actual'], colnames=['Predicted'])
cm_log
```

```python
xgb_accuracy = (cm_xgb.iloc[0, 0] + cm_xgb.iloc[1, 1])/(cm_xgb.values.sum())
rf_accuracy = (cm_rf.iloc[0, 0] + cm_rf.iloc[1, 1])/(cm_rf.values.sum())
log_accuracy = (cm_log.iloc[0, 0] + cm_log.iloc[1, 1])/(cm_log.values.sum())

{
    "xgb": xgb_accuracy, 
    "rf": rf_accuracy, 
    "log": log_accuracy
}
```

Compare log loss

```python
from sklearn.metrics import log_loss
```

```python
xgb_logloss = log_loss(df_accuracy["game_next_goal"], df_accuracy["xgb_prediction_prob"])
rf_logloss = log_loss(df_accuracy["game_next_goal"], df_accuracy["rf_prediction_prob"])
log_logloss = log_loss(df_accuracy["game_next_goal"], df_accuracy["log_prediction_prob"])

{
    "xgb": xgb_logloss, 
    "rf": rf_logloss, 
    "log": log_logloss
}
```

Compare AUC

```python
from sklearn.metrics import roc_curve, auc
```

```python
fpr, tpr, threshold = roc_curve(df_accuracy["game_next_goal"], df_accuracy["xgb_prediction"])
xgb_auc = auc(fpr, tpr)


fpr, tpr, threshold = roc_curve(df_accuracy["game_next_goal"], df_accuracy["rf_prediction"])
rf_auc = auc(fpr, tpr)

fpr, tpr, threshold = roc_curve(df_accuracy["game_next_goal"], df_accuracy["log_prediction"])
log_auc = auc(fpr, tpr)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(fpr, tpr, alpha = 1, label = f"XGBoost Model AUC: {xgb_auc:.3f}", linewidth = 3)
plt.plot(fpr, tpr, alpha = 0.55, label = f"Random Forest Model AUC: {rf_auc:.3f}", linewidth = 7)
plt.plot(fpr, tpr, alpha = 0.1, label = f"Logistic Model AUC: {log_auc:.3f}", linewidth = 11)

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.savefig('AUC_plot.png')
```

The lines are so close to each other that they are indistinguishable on the graph

Overall Comparison

```python
df_compare = pd.DataFrame({ 
                           "Accuracy": [xgb_accuracy, rf_accuracy, log_accuracy], 
                           "AUC": [xgb_auc, rf_auc, log_auc], 
                           "Log Loss": [xgb_logloss, rf_logloss, log_logloss]
                          })
df_compare.index = ["XGBoost", "Random Forest", "Logistic"]
df_compare
```
