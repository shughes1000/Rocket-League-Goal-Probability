```python
from sklearn.calibration import calibration_curve
```

```python
?calibration_curve
```

```python
y_means_orange, proba_means_orange = calibration_curve(df_holdout.loc[df_holdout["prediction_prob"] >= 0.5, "game_next_goal"], 
                                                       df_holdout.loc[df_holdout["prediction_prob"] >= 0.5, "prediction_prob"], 
                                                       n_bins = 50, strategy = "quantile")
```

```python
proba_means_orange
```

```python
y_means_blue, proba_means_blue = calibration_curve(df_holdout.loc[df_holdout["prediction_prob"] < 0.5, "game_next_goal"], 
                                                     df_holdout.loc[df_holdout["prediction_prob"] < 0.5, "prediction_prob"], 
                                                     n_bins = 50, strategy = "quantile")
```

```python
proba_means_blue
```

```python
def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return idx/100
```

```python
df_holdout["prediction_prob_adj"] = 0
df_holdout.head()
```

```python
df_holdout.loc[df_holdout["prediction_prob"] < 0.5, "prediction_prob_adj"] = 0.01 + df_holdout.loc[df_holdout["prediction_prob"] < 0.5, "prediction_prob"].apply(lambda x: find_nearest(proba_means_blue, x))
df_holdout.loc[df_holdout["prediction_prob"] >= 0.5, "prediction_prob_adj"] = 0.5 + df_holdout.loc[df_holdout["prediction_prob"] >= 0.5, "prediction_prob"].apply(lambda x: find_nearest(proba_means_orange, x))
df_holdout.sample(15)
```

```python
df_holdout["prediction_prob_adj"].hist()
```

```python
df_grouped = df_holdout[["game_next_goal", "prediction_prob", "prediction_prob_adj"]].groupby(round(df_holdout["prediction_prob"], 1)).mean()
```
