```python
import pandas as pd
%config InlineBackend.figure_formats = ['svg']
```

```python
df_train = pd.read_pickle("training_data_final.pkl")
df_train.head()
```

```python
df_train.sample(15)
```

```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree
from sklearn.metrics import balanced_accuracy_score, roc_auc_score, make_scorer
from sklearn.metrics import confusion_matrix, plot_confusion_matrix
```

```python
# FIX THIS LATER
# I am Removing instances of demos with this code I believe
df_train = df_train.loc[df_train["blue_pos_x"].isnull() == False, ]
df_train = df_train.loc[df_train["blue_boost"].isnull() == False, ]
df_train = df_train.loc[df_train["orange_pos_x"].isnull() == False, ]
df_train = df_train.loc[df_train["orange_boost"].isnull() == False, ]
```

```python
X = df_train[["blue_pos_x", "blue_pos_y", "blue_pos_z", "blue_boost", 
              "orange_pos_x", "orange_pos_y", "orange_pos_z", "orange_boost", 
              "ball_pos_x", "ball_pos_y", "ball_pos_z"]]
X.head()
```

```python
y = df_train["game_next_goal"].copy()
y.head()
```

```python
rf = RandomForestClassifier(random_state = 99, max_depth = 32, verbose = 3)
rf.fit(X, y)
```

```python
rf.score(X, y)
```

```python
df_holdout = pd.read_pickle("holdout_data_final.pkl")
```

```python
# FIX THIS LATER
# I am Removing instances of demos with this code I believe
df_holdout = df_holdout.loc[df_holdout["blue_pos_x"].isnull() == False, ]
df_holdout = df_holdout.loc[df_holdout["blue_boost"].isnull() == False, ]
df_holdout = df_holdout.loc[df_holdout["orange_pos_x"].isnull() == False, ]
df_holdout = df_holdout.loc[df_holdout["orange_boost"].isnull() == False, ]
```

```python
X_holdout = df_holdout[["blue_pos_x", "blue_pos_y", "blue_pos_z", "blue_boost", "orange_pos_x", "orange_pos_y", "orange_pos_z", "orange_boost", "ball_pos_x", "ball_pos_y", "ball_pos_z"]]
y_holdout = np.array(df_holdout["game_next_goal"])
X_holdout.info()
```

```python
rf.score(X_holdout, y_holdout)
```

```python
df_holdout["prediction"] = rf.predict(X_holdout)
```

```python
cm = pd.crosstab(df_holdout['game_next_goal'], df_holdout['prediction'],
                               rownames=['Actual'], colnames=['Predicted'])
cm
```

```python
import seaborn as sn
sn.heatmap(cm, annot=True, cmap = 'Blues', fmt = 'g')
```

```python
df_holdout["prediction_prob"] = rf.predict_proba(X_holdout)[:, 1]
```
