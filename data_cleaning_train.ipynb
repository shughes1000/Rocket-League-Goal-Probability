# Cleaning the Training Data

In this file, I take the data and clean it so it will be ready for model building

```python
import pandas as pd
```

Load in datasets created by data_setup

```python
df_train = pd.read_csv("training_data1.csv", index_col = 0)
```

```python
df_train.shape
```

```python
df_train = df_train.append(pd.read_csv("training_data2.csv", index_col = 0))
```

```python
df_train.shape
```

```python
df_train = df_train.append(pd.read_csv("training_data3.csv", index_col = 0))
```

```python
df_train.shape
```

The indeces overlapped so I had to reset them

```python
df_train.reset_index(drop = True, inplace = True)
```

I am only interested in the who scores the next goal so we can remove any instance where another goal does not occur

```python
df_train = df_train.loc[df_train["game_next_goal"].isnull() == False, ]
```

```python
df_train.info()
```

```python
for col in df_train.columns:
    if df_train[col].dtype == "object":
        print(f"{col}: {df_train[col].unique()}")
        print()
```

A lot of the columns were not the correct type. There were also some issues with booleans as string. The following code fixes all of this. 

```python
df_train["blue_handbrake"] = df_train["blue_handbrake"].astype("bool")
df_train["orange_handbrake"] = df_train["orange_handbrake"].astype("bool")

df_train["blue_ball_cam"] = df_train["blue_ball_cam"].astype("bool")
df_train["orange_ball_cam"] = df_train["orange_ball_cam"].astype("bool")

df_train["blue_boost_active"] = df_train["blue_boost_active"].astype("bool")
df_train["orange_boost_active"] = df_train["orange_boost_active"].astype("bool")
```

```python
df_train.loc[df_train["game_is_overtime"] == "True", "game_is_overtime"] = True
df_train.loc[df_train["game_is_overtime"].isnull(), "game_is_overtime"] = False
df_train.loc[df_train["game_is_overtime"] == "no_OT", "game_is_overtime"] = False
df_train["game_is_overtime"] = df_train["game_is_overtime"].astype("bool")

df_train["game_ball_has_been_hit"] = df_train["game_ball_has_been_hit"].astype("bool")
```

```python
df_train.loc[df_train["blue_dodge_active"]== "False", "blue_dodge_active"] = 0
df_train.loc[df_train["orange_dodge_active"] == "False", "orange_dodge_active"] = 0

df_train["blue_dodge_active"] = df_train["blue_dodge_active"].astype("float64")
df_train["orange_dodge_active"] = df_train["orange_dodge_active"].astype("float64")
```

```python
df_train.loc[df_train["blue_jump_active"]== "False", "blue_jump_active"] = 0
df_train.loc[df_train["orange_jump_active"] == "False", "orange_jump_active"] = 0

df_train["blue_jump_active"] = df_train["blue_jump_active"].astype("float64")
df_train["orange_jump_active"] = df_train["orange_jump_active"].astype("float64")
```

```python
df_train.loc[df_train["blue_double_jump_active"]== "False", "blue_double_jump_active"] = 0
df_train.loc[df_train["orange_double_jump_active"] == "False", "orange_double_jump_active"] = 0

df_train["blue_double_jump_active"] = df_train["blue_double_jump_active"].astype("float64")
df_train["orange_double_jump_active"] = df_train["orange_double_jump_active"].astype("float64")
```

```python
df_train.info()
```

```python
for col in df_train.columns:
    if df_train[col].dtype == "bool":
        print(f"{col}: {df_train[col].unique()}")
        print()
```

```python
df_train.sample(15)
```

Now the data is ready for model building, so we can store it as a pickle file

```python
df_train.to_pickle("training_data_final.pkl")
```
