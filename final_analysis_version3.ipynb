### Feature Importance

```python
df_features = pd.DataFrame({"Feature": X_accuracy.columns, 
                            "Importance": clf_rf.feature_importances_.round(3)})
df_features.sort_values(by = "Importance", axis = 0, ascending = False, ignore_index = True, inplace = True)
df_features
```

Based on our feature importance, we see that the ball's angular velocity and each players' distance to the ball are among the most important features in our model. 

### Visualizations

```python
df_accuracy["prediction_prob"] = clf_rf.predict_proba(X_accuracy)[:, 1]
df_accuracy["prediction_prob"].hist()
```

```python
df_accuracy["correct_prediction"] = 0
df_accuracy.loc[df_accuracy["game_next_goal"] == df_accuracy["prediction"], "correct_prediction"] = 1
df_accuracy.sample(5)
```

```python
df_grouped = df_accuracy.groupby("time_to_next_goal").agg("correct_prediction").describe()
df_grouped.reset_index(inplace = True)
df_grouped
```

```python
from plotnine import *
```

```python
(
 ggplot(df_grouped, aes(x = "time_to_next_goal", y = "mean"))
 + geom_col(aes(fill = "count"), color = "white")
 + labs(title = "Model Accuracy by Time Until Next Goal",
        x = "Seconds Until Next Goal", y = "Model Accuracy")
 + scale_fill_gradient(low = "lightgreen", high = "darkgreen")
 + scale_x_continuous(breaks = [i for i in range(0, 131, 10)])
 + scale_y_continuous(breaks = [i/10 for i in range(11)])
 + theme_bw()
 + theme(figure_size = (16, 6), title = element_text(lineheight = 1.1))
)
```

Same thing but for time as less than or equal

```python
df_accuracy["time_to_next_goal"].describe()
```

```python
df_lt = pd.DataFrame({"time_to_next_goal": [i for i in range(125)], 
                      "count": [df_accuracy.loc[df_accuracy["time_to_next_goal"] <= i, 
                                                "correct_prediction"].count() for i in range(125)], 
                      "mean": [df_accuracy.loc[df_accuracy["time_to_next_goal"] <= i, 
                                               "correct_prediction"].mean() for i in range(125)]})
df_lt.head()
```

```python
p = (
     ggplot(df_lt, aes(x = "time_to_next_goal", y = "mean"))
     + geom_col(aes(fill = "count"), color = "white")
     + labs(title = "Model Accuracy by Time Until Next Goal (Cumulative)",
            x = "Maximum Amount of Seconds Until Next Goal", y = "Model Accuracy")
     + scale_fill_gradient(low = "lightblue", high = "darkblue")
     + scale_x_continuous(breaks = [i for i in range(0, 131, 10)])
     + scale_y_continuous(breaks = [i/10 for i in range(11)])
     + theme_bw()
     + theme(figure_size = (8, 3), title = element_text(lineheight = 1.1))
    )

p.save("cumulative_accuracy.png")
p
```

```python
df_analysis_graph = df_analysis.loc[df_analysis["game_is_overtime"] == True,
                                   ].groupby("game_seconds_remaining")["prediction_prob"].describe().reset_index()
df_analysis_graph.head()
```

```python
df_analysis_graph["prediction"] = 0
df_analysis_graph.loc[df_analysis_graph["mean"] >= 0.5, "prediction"] = 1
df_analysis_graph["prediction"] = df_analysis_graph["prediction"].astype("category")
```

Graph of overtime period (note that the probability for each second is aggregated)

```python
p = (
     ggplot(df_analysis_graph, 
            aes(x = "game_seconds_remaining", y = "mean"))
     + geom_col(aes(fill = "prediction"), color = "white")
     + labs(title = "Model Prediction",
            x = "Seconds", y = "Prediciton")
     + scale_fill_manual(["blue", "darkorange"])
     + scale_x_continuous(breaks = [i*5 for i in range(18)])
     + scale_y_continuous(breaks = [i/10 for i in range(11)])
     + theme_bw()
     + theme(figure_size = (8, 3), title = element_text(lineheight = 1.1))
    )

p.save("real_game.png")
p
```
